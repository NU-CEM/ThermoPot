```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
BZS = []
BaS = []
ZrS2 =[]
Ba =[]
Zr =[]
S = []
temp=[]

with open('/Users/prakriti/ThermoPot/BaZrS3/phonopy_output/BaZrS3_Pnma.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            BZS.append(float(line[2]))
with open('/Users/prakriti/ThermoPot/BaZrS3/phonopy_output/BaS_Fm-3m.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            BaS.append(float(line[2]))
with open('/Users/prakriti/ThermoPot/BaZrS3/phonopy_output/ZrS2_P-3m1.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            ZrS2.append(float(line[2]))
with open('/Users/prakriti/ThermoPot/BaZrS3/phonopy_output/Ba_Im-3m.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            Ba.append(float(line[2]))
with open('/Users/prakriti/ThermoPot/BaZrS3/phonopy_output/S_P2_c.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            Zr.append(float(line[2]))
with open('/Users/prakriti/ThermoPot/BaZrS3/phonopy_output/Zr_P63_mmc.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            temp.append(float(line[0]))
            S.append(float(line[2]))

```

```python
binary_sulfides = np.array(BZS)-np.array(BaS)-np.array(ZrS2)
elemental = np.array(BZS)-np.array(Ba)-np.array(Zr)-3*np.array(S)
```

```python
plt.plot(temp,binary_sulfides, label='BaS + ZrS$_2$ --> BaZrS$_3$')
plt.plot(temp,elemental, label='Ba + Z r+ 3S --> BaZrS$_3$')
plt.legend()
plt.ylabel('S (kJ/mol)')
plt.xlabel('Temperature (K)')
plt.savefig('entropy_contributions.png')
```
