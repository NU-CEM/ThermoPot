```python
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
temperature = []
BZS_GFE = []
Ba4Zr3S10_GFE = []
Ba2ZrS4_GFE = []
ZrS2_GFE = []
with open('BZS_4310_214_ZrS2') as file:
    for line in file:
        line = line.split()
        temperature.append(float(line[0]))
        temperature1 = np.array(temperature)
        BZS_GFE.append(float(line[1]))
        BZS_GFE1 = np.array(BZS_GFE)
        BZS_GFE1 = BZS_GFE1/20/5
        Ba4Zr3S10_GFE.append(float(line[3]))
        Ba4Zr3S10_GFE1 = np.array(Ba4Zr3S10_GFE)
        Ba4Zr3S10_GFE1 = Ba4Zr3S10_GFE1/17
        Ba2ZrS4_GFE.append(float(line[5]))
        Ba2ZrS4_GFE1 = np.array(Ba2ZrS4_GFE)
        Ba2ZrS4_GFE1 = Ba2ZrS4_GFE1/7
        ZrS2_GFE.append(float(line[7]))
        ZrS2_GFE1 = np.array(ZrS2_GFE)
        ZrS2_GFE1 = ZrS2_GFE1/3
#print(BZS_GFE1[:5])
#print(Ba2ZrS4_GFE1[:5])
#print(ZrS2_GFE1[:5])
```

```python
reaction_214 = (2*(BZS_GFE1) - Ba2ZrS4_GFE1 - ZrS2_GFE1)/2
print(reaction_214)
reaction_4310 = (4*(BZS_GFE1) - Ba4Zr3S10_GFE1 - ZrS2_GFE1)/4
print(reaction_4310)
```

```python
plt.plot(temperature1,reaction_214,"--", "")
plt.plot(temperature1,reaction_4310,".")
plt.ylabel(r"$\Delta G$ (kJ mol$^{-1}$)")
#plt.xlim([200,1300])
#plt.ylim([-8893,-8892.5])
#plt.yticks([-8893,-8892.5])
plt.xlabel("Temperature (K)")
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('GFE_ternary_phases')
```
